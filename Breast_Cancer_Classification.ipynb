- Breast cancer is a kind of cancer that begins with the growth of cells in the breast tissue. After skin cancer, breast cancer is the most common cancer diagnosed in women in the United States. But breast cancer doesn't just happen in women. Everyone is born with some breast tissue, so anyone can get breast cancer. Breast cancer survival rates have been increasing. Signs of breast cancer may include a lump in the breast, a change in breast shape, dimpling of the skin, milk rejection, fluid coming from the nipple, a newly inverted nipple, or a red or scaly patch of skin.
```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

*Types Of Tumor We Are Working On- Benign and Malignant*


**Data Collection & Processing**
- loading the data from sklearn

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

- Loading the data to a data frame





```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

-**Work Flow**:

1.   *DataSet(data collection)*
2.   *Data Preprocessing*
3.   *Train Test Split*
4.   *Logistic Regression(Machine Learning Model)*



```python
data_frame.head() #first rows
```

- adding the 'target' column to the data frame


```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail() #last rows
```

```python
data_frame.shape
```

- some information about the data

```python
data_frame.info()
```

```python
data_frame.isnull().sum() #checking missing values
```

```python
data_frame.describe()
```

- checking the distribution of target variable

```python
data_frame['label'].value_counts()
```

1 -> **Benign**

0 -> **Malignant**

```python
data_frame.groupby('label').mean()
```

- separating the features and target


```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

- splitting the data into training data & testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING - ***LOGISTIC REGRESSION***

```python
model = LogisticRegression()
```

- training the Logistic Regression model using training data

```python
model.fit(X_train, Y_train)
```

Model Evaluation - Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```
